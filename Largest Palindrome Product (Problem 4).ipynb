```python
def LargestPalindromeProduct(maxNum):
    found = False
    largest = 0
    
    for i in range(maxNum, 10, -1):
        for j in range(maxNum, 10, -1):
            if str(i*j) == str(i*j)[::-1] and i*j > largest:
                largest = i*j
    return largest
```

```python
print(LargestPalindromeProduct(99))
print(LargestPalindromeProduct(999))
```
