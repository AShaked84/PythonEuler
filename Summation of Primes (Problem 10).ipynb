```python
def seive(maxNum):
    primes = [True] * (maxNum+1)
    i = 2

    while i*i <= maxNum:
        if primes[i]:
            for j in range(i*i, maxNum+1, i):
                primes[j] = False
        i+=1
    return primes
```

```python
maxPrime = 2000000
sum = 0

primes = seive(maxPrime)

for i in range(2, maxPrime+1):
    if primes[i]:
        sum += i

print(sum)
```
