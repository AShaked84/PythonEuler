```python
def isPrime(n):
    if n <= 1:
        return False

    i = 2
    while i*i <= n:
        if n%i == 0:
            return False
        i +=1
    return True
```

```python
def NthPrime(num):
    i = 0
    counter = 0
    while counter < num:
        i+=1
        if isPrime(i):
            counter+=1

    return i
```

```python
print(NthPrime(6))
print(NthPrime(10001))
```
