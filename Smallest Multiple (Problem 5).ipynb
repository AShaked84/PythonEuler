```python
def GCD(a, b):
    while b != 0:
        temp = b
        b = a%b
        a = temp
    return a
```

```python
def SmallestMultiple(numberList):
    lcm = 1

    for i in numberList:
        gcd = GCD(i, lcm)
        lcm = (lcm * i) / gcd

    return lcm
```

```python
print(SmallestMultiple(range(1, 10)))
print(SmallestMultiple(range(1, 20)))
```
