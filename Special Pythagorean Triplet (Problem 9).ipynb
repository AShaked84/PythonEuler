```python
def triplets(mysum):
    m = 2; n = 1

    while m < 100:
        n = 1
        while n < m:
            a = m*m - n*n
            b = 2*n*m
            c = n*n + m*m

            if sum([a, b, c]) == mysum:
                return a*b*c
            else: 
                n += 1
        m += 1
```

```python
print(triplets(1000))
```
