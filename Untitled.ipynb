```python
def SpecialPythagoreanTriangle(sideSum):
    search = True
    m = 2
    n = 1

    a = b = c = 0
    while search:
        while n < m:
            a = m*m - n*n
            b = 2*m*n
            c = m*m + n*n

            if a + b + c == sideSum:
                search = False
                break
            n += 1
        m += 1

    return (a, b, c)
```

```python
print(SpecialPythagoreanTriangle(1000))
```
