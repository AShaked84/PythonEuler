```python
import math
```

```python
def split_number_string(number):
  s = str(number)
  if s.startswith('-'):
    # Include a minus sign in the result or just the digits
    return [int(d) for d in s[1:]]
    # To include the negative sign with the first digit: 
    # digits = [int(d) for d in s[1:]]
    # digits[0] = -digits[0]
    # return digits
  else:
    return [int(d) for d in s]
```

```python
def LargestProduct(number, windowLen):
    numList = split_number_string(number)
    largestWin = 0

    i = 0
    while i <= len(numList)-windowLen:
        current_window = numList[i:i + windowLen]
        
        if 0 in current_window:
            i += current_window.index(0) + 1
            continue

        window = math.prod(current_window)
        
        if window > largestWin:
            largestWin = window

        i += 1

    return largestWin
```
