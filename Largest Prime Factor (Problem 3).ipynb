```python
def largestPrimeFactor(num):
    factor = 2  # start with the smallest prime

    # divide out all factors up to sqrt(num)
    while factor * factor <= num:
        if num % factor == 0:
            num //= factor  # remove this factor
        else:
            factor += 1 if factor == 2 else 2  # after 2, check only odd numbers

    # whatever is left is the largest prime factor
    return num

# Test with the example number
num = 600851475143
print("Largest prime factor:", largestPrimeFactor(num))
```
